```python
data = pd.read_csv("train.csv")
data.head()
```

```python
df = data.isna().sum().sort_values(ascending=False)
```

```python
alempty = []
for i in range(len(df)):
    if df[i]>len(data)/20:
        alempty.append(df.index[i])
```

```python
alempty
```

```python
tobe = data[alempty]
tobe
data.drop(alempty,axis=1,inplace=True)
```

```python
for label ,content in data.items():
    if pd.api.types.is_string_dtype(content):
        data[label] = content.astype("category").cat.as_ordered()
```

```python
for label,content in data.items():
    if pd.api.types.is_categorical_dtype(content):
        data[label] = pd.Categorical(content).codes+1
        data[label+"_is_missing"] = pd.isnull(content)
    elif pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            data[label+'_is_missing'] = pd.isnull(content)
            data[label] = content.fillna(content.median())
```

```python
data.info()
```

```python
data.head()
```

```python
kmeans = KMeans(n_clusters=8)
```

```python
kmeans.fit(data)
```

```python
colormap = np.array(['red','green','blue','salmon','grey','yellow','pink'])
fig,ax = plt.subplots(figsize=(12,7))
ax.scatter(x=data.LotArea,y=data.SalePrice,c=colormap[kmeans.labels_-1])
```

```python
labels = pd.Series(kmeans.labels_,name="labels")
```

```python
group = pd.concat([tobe,labels],names=['labels'],axis=1)
```

```python
for i in range(len(tobe)):
    for row in group[alempty[i]]:
        row = group[alempty[i]][group['labels'][]].median()
```

```python
for i in range(8):
    temp = group[group.labels==i]
    for label,content in temp.items():
        if pd.api.types.is_numeric_dtype(content):
            if pd.isnull(content).sum():
                if content.median():
                    group[label] = content.fillna(group[label].median())
                else :
                    group[label] = content.fillna(content.median())
```
